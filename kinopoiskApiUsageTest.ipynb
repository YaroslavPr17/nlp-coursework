```python
# Source: https://python-scripts.com/requests#auth-requests-python

import requests

from requests.auth import AuthBase
 
class TokenAuth(AuthBase):
    """Implements a custom authentication scheme."""
 
    def __init__(self, token):
        self.token = token
 
    def __call__(self, r):
        """Attach an API token to a custom auth header."""
        r.headers['X-TokenAuth'] = f'{self.token}'  # Python 3.6+
        return r
```

```python
BASE_URL = 'https://kinopoiskapiunofficial.tech'
response = requests.get(f"{BASE_URL}/api/v2.2/films/405609/reviews?page=51&order=DATE_DESC", headers={"X-API-KEY": "61c2d911-e649-4afc-9790-676eedf367ee"})
res = response.json().get("items")
```
